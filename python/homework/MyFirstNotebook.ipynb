```python
from skimage import io
```

```python
img = io.imread('./3.jpg')
plt.figure()
plt.imshow(img)
plt.show()
```

```python
plt.figure()
plt.imshow(img[:20, :20, 2], interpolation='nearest')
plt.show()
```

```python
from skimage import filters
filtered_img = filters.gaussian(img, 60)
plt.imshow(filtered_img)
plt.show()
```
